```python
#Word Embedding Techniques using Embedding Layer in Keras
### Libraries USed Tensorflow> 2.0  and keras
##tensorflow >2.0
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#we need to convert these sentences in one hot
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent


```

```python
### Vocabulary size
#meaning what is the size of the dictionary
voc_size=10000

```

```python
#One Hot Representation
onehot_repr=[one_hot(words,voc_size)for words in sent] 
#i am going to pass this word to one hot function where i have to pass word and other is vocsize
#with this we will be getting index from the dictionary
print(onehot_repr)

#the index for 'the' is 7117 out of the 10000 voc size
#index for glass is 318 and so on
#when u see the first and second sentence ,they are almost same, and their indexes also matches
```

```python
#Word Embedding Represntation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences#
from tensorflow.keras.models import Sequential


```

```python
import numpy as np

```

```python
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)

```

```python

dim=10

#dimension mean features

```

```python

model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')


```

```python
model.summary()


```

```python
print(model.predict(embedded_docs))

```
